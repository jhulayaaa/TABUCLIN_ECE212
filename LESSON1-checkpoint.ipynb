> ### Exercise 4: Area and Volume (Solved, 15 Lines)

```python
import math
radius = int(input('Please enter your radius in cm: '))
area = (radius**2)*(math.pi)
volume = (4/3)*(math.pi)*(radius**3)
print('Your area is:')
print( area, 'cm squared')
print('Your volume is:')
print( volume, 'cm cubed')
```
