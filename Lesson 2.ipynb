```python
import math
g1 = int(input('Please enter the first longitude in km: '))
g2 = int(input('Please enter the second longitude in km: '))
t1 = int(input('Please enter the first latitude in km: '))
t2 = int(input('Please enter the second latitude in km: '))
r = 6371.01
g = g1-g2
d = r*(math.acos(((math.sin(t1))*(math.sin(t2)))+((math.cos(t1))*(math.cos(t2))*(math.cos(g)))))
print('The distance is:', d, 'km')
```
